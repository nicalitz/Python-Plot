## Example: Two plots on the same axes

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generate data to plot
t = np.arange(0.01, 10.0, 0.01)
s1 = np.exp(t)
s2 = np.sin(2 * np.pi * t)
```

```python
# define figure and primary axes
fig, ax1 = plt.subplots(figsize=(12,4))

# define second axes using same x-axis
ax2 = ax1.twinx()

# plot data to axes
ax1.plot(t, s1, color='C0')
ax2.plot(t, s2, color='C1', linewidth=0.75)

# define grid
ax1.grid(linewidth=0.5, alpha=0.5, linestyle='dotted')

# color ticklabels based on plots
ax1.set_ylabel('s1 data', color='C0')
ax1.tick_params('y', colors='C0')   # note 'colors' used instead of 'color'
ax2.set_ylabel('s2 data', color='C1')
ax2.tick_params('y', colors='C1')

# define x-axis label
ax1.set_xlabel('iteration')

# could also color as follow (color every second label):
for index, label in enumerate(ax2.get_yticklabels()):
    if (index % 2):
        label.set_color('C2')
        label.set_alpha(0.4)

plt.show()
```
